This notebook is a template for downloading an analysis-ready dataset of global COVID-19 confirmed cases and deaths by the State-level and for some countries at the Adminstrative division 1 (admin1)-level (states, provinces, etc.)

Data are presented in two formats
1. Original JHU format with each day as a separate column
2. Time series format with cummulative confirmed cases and death by day

Data source: [COVID-19 Data Repository by the Center for Systems Science and Engineering (CSSE) at Johns Hopkins University](https://github.com/CSSEGISandData/COVID-19)

Author: Peter Rose (pwrose@ucsd.edu)

```python
import os
import pandas as pd
import dateutil
```

```python
pd.options.display.max_rows = None  # display all rows
pd.options.display.max_columns = None  # display all columsns
```

### Process global cummulative confirmed cases

```python
confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
confirmed.fillna('', inplace=True)
# use consistent column names
confirmed.rename(columns={'Province/State': 'admin1', 'Country/Region': 'country',
                          'Lat': 'lat', 'Long': 'long'}, inplace=True)
```

```python
confirmed.head()
```

```python
confirmed.query("admin1 != ''")
```

#### Reformat dataframe by day

```python
days = confirmed.columns.tolist()[4:]
```

```python
def split_by_day(df, day, label):
    day_df = df[['country', 'admin1','lat', 'long', day]].copy()
    day_df.rename(columns={day: label}, inplace=True)
    day_df['date'] = day
    return day_df
```

```python
timeseries_confirmed = pd.concat((split_by_day(confirmed, day, 'cases') for day in days))
timeseries_confirmed['date'] = timeseries_confirmed['date'].apply(dateutil.parser.parse)
```

### Process global cummulative deaths

```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
```

```python
deaths.fillna('', inplace=True)
deaths.rename(columns={'Province/State': 'admin1', 'Country/Region': 'country',
                       'Lat': 'lat', 'Long': 'long'}, inplace=True)
```

```python
timeseries_deaths = pd.concat((split_by_day(deaths, day, 'deaths') for day in days))
timeseries_deaths['date'] = timeseries_deaths['date'].apply(dateutil.parser.parse)
```

```python
timeseries_deaths.shape
```

```python
timeseries_deaths.tail()
```

### Merge global cases and deaths into a timeseries dataframe

```python
# remove Lat and Long to avoid merging on floating point numbers
subset = timeseries_confirmed.drop(['lat','long'], axis=1)
timeseries = subset.merge(timeseries_deaths, on=['country', 'admin1', 'date'])
```
